## Check for columns that describe the same data
- count distinct entries of each pair of columns

```python
unique_count = """
    select 
        count(distinct dest_country) as count_dest_country,
        count(distinct dest_country_name) as count_dest_country_name,
        count(distinct carrier_group) as count_carrier_group,
        count(distinct carrier_group_new) as count_carrier_group_new,
        count(distinct origin_airport_id) as count_origin_airport_id,
        count(distinct origin) as count_origin,
        count(distinct origin_country) as count_origin_country,
        count(distinct origin_country_name) as count_origin_country_name,
        count(distinct dest_airport_id) as count_dest_airport_id,
        count(distinct dest) as count_dest,
        count(distinct unique_carrier) as count_unique_carrier,
        count(distinct unique_carrier_name) as count_unique_carrier_name,
        count(distinct carrier) as count_carrier,
        count(distinct carrier_name) as count_carrier_name
    from passengers
"""
res = execute_read_query(con, unique_count)
```

## Check the distinct value pairs of origin airport id and origin 

count_origin_airport_id	2513  
count_origin	2515  

```python
unique_airport_origin = """
    select 
        distinct origin_airport_id, origin
    from passengers
"""
res = execute_read_query(con, unique_airport_origin)
res
```

#### Check for null values between airport origin and origin

```python
res[res.isna().any(axis=1)]
```

#### Count how many origin_airport_id records have duplicate origin values

```python
res_count = res.groupby('origin_airport_id').agg(
    count=('origin','count')).reset_index().sort_values('count', ascending=False)
res_count
```

```python
find_rows = res[res['origin_airport_id'].isin(['12544', '13788','12395','15622'])]
find_rows
```

#### Conclusion: 
- drop origin because it changes over time, but origin_airport_id is the same
